```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
archivo = './StudentPerformanceFactors.csv'
student_performance = pd.read_csv(archivo)
student_performance.info()
```

```python
student_performance.head()
```

```python
student_performance.describe()
```

```python
conteo = student_performance['Teacher_Quality'].value_counts()
plt.bar(conteo.index,conteo.values,color='brown')
plt.title('Calidad del profesor')
plt.ylabel('Universo de alumnos')
for indice,valor in(enumerate(conteo.values)):
    plt.text(indice,valor+1,str(valor),ha='center',fontsize=10)
plt.show()
```

```python
conteo = student_performance['School_Type'].value_counts()
plt.bar(conteo.index,conteo.values,color='brown')
plt.title('Tipo de escuela')
plt.ylabel('Universo de alumnos')
for indice,valor in(enumerate(conteo.values)):
    plt.text(indice,valor+1,str(valor),ha='center',fontsize=10)
plt.show()
```

```python
conteo = student_performance['Parental_Education_Level'].value_counts()
plt.bar(conteo.index,conteo.values,color='brown')
plt.title('Nivel educativo de los padres')
plt.ylabel('Universo de alumnos')
for indice,valor in(enumerate(conteo.values)):
    plt.text(indice,valor+1,str(valor),ha='center',fontsize=10)
plt.show()
```

```python
conteo = student_performance['Distance_from_Home'].value_counts()
plt.bar(conteo.index,conteo.values,color='brown')
plt.title('Nivel educativo de los padres')
plt.ylabel('Universo de alumnos')
for indice,valor in(enumerate(conteo.values)):
    plt.text(indice,valor+1,str(valor),ha='center',fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=student_performance['Exam_Score'],y=student_performance['Hours_Studied'],color='brown',alpha=0.45)
plt.xlabel('Puntaje del examen')
plt.ylabel('Horas de estudio')
plt.grid(visible=1, linestyle='--',alpha=0.30)
plt.show()
```
